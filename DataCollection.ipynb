```python
import json

path = "/content/drive/MyDrive/RTRP/data/train_data.json"

with open(path, "r", encoding="utf-8") as f:
    outer_list = json.load(f)

print("Data loaded successfully")
print("Total records:", len(outer_list))
```

```python
corpus = []

for item in outer_list:
    inner_dict = json.loads(item)

    text_block = ""
    for key, value in inner_dict.items():
        text_block += f"{key}: {value}\n"

    corpus.append(text_block.strip())

output_path = "/content/drive/MyDrive/RTRP/data/corpus.txt"

with open(output_path, "w", encoding="utf-8") as f:
    f.write("\n\n".join(corpus))

print("Corpus created successfully")
```
